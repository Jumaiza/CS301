```python
import numpy as np
from numpy import mean,random
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot
from numpy import cov
from scipy.stats import pearsonr, chi2_contingency
from scipy.stats import spearmanr
import scipy.stats as stats
import pandas as pd
import seaborn as sns
import statistics
```

```python
#QUESTION 2 #1 Sample Size 2
seed(10)
normaldist2=random.normal(loc=41.2,scale=12.56,size=2)
mean2=[]
for i in range(200):
    mean=np.random.choice(normaldist2,2).mean()
    mean2.append(mean)

print("Mean: ",statistics.mean(mean2))
print("Standard Deviation/Error: ",statistics.stdev(mean2))
pyplot.hist(mean2);
```

The mean of the sample means of age seems to decrease as the sameple size is increase. This is proved by the central limit theorem which states that as the sample size increase, the mean of the sampling distribution tends to get closer to the true sample mean which is 41.2.

The standard deviation/error decreases as the sample size is increased. This is because as sample size increases, the sampling distribuition of the mean is approaching a normal distribution. This can be shown by the histograms that are plotted.
