```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules, fpmax
import matplotlib.pyplot as plt
```

```python
print("QUESTION A:")

df = pd.read_csv('Online Retail.csv')

df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]

cleanup = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

def encode_values(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

encoded_values = cleanup.applymap(encode_values)
encoded_values.drop('POSTAGE', inplace=True, axis=1)

freq_items = apriori(encoded_values, min_support=0.03, use_colnames=True)
freq_items
freq_items.to_csv('yo.csv')
```

The frequent itemsets for the dataset with support of 3% are displayed. We can see that most itemsets have a support value of around 4%.

```python
print("QUESTION B:")
rules = association_rules(freq_items, metric="confidence", min_threshold=0.3)
rules
```

The association rules for the frequent itemsets with 30% confidence are displayed. Based on the highest lift values, items in row 2-3 and the last 4 rows have the strongest rules. Most rules have over 50% confidence which is great.

```python
print("QUESTION C:")
maxItems = fpmax(encoded_values, min_support=0.03,use_colnames=True).sort_values(by=['support'],ascending=False)
maxItems
```

The maximum itemsets for the frequent items in (A) are displayed. This means there is 114 itemsets which are frequent but none of their supersets are frequent.
